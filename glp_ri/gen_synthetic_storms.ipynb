```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
```

```python
synth_storm = np.zeros((380, 540), dtype='float')
x = np.arange(0, 540)
y = np.arange(0, 380)[:, None]
```

```python
r = 30
x0 = 100
y0 = 100
circle = ((x - x0)**2 + (y - y0)**2 <= r**2).astype('float')
```

```python
plt.imshow(circle)
```

```python
e = 30
x0 = 100
y0 = 100
square = np.logical_and(np.abs(x - x0) <= e, np.abs(y - y0) <= e).astype('float')
```

```python
plt.imshow(square)
```
